```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

pd.options.display.max_rows=500
pd.options.display.max_columns=50
```

```python
train = pd.read_feather('../data/input/train_data.feather')
test = pd.read_feather('../data/input/test_data.feather')
```

```python
train.head()
```

```python
train.describe().T
```

```python
test.describe().T
```

# position

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
sns.countplot(train['position'], ax=ax1)
sns.countplot(test['position'], ax=ax2)
plt.tight_layout()
```

# age

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
sns.distplot(train['age'], ax=ax1)
sns.distplot(test['age'], ax=ax2)
plt.tight_layout()
```

# area

```python
order = train['area'].unique()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 50))
sns.countplot(data=train, y='area', order=order, ax=ax1)
sns.countplot(data=test, y='area', order=order, ax=ax2)
plt.tight_layout()
```
