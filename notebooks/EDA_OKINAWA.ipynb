```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

pd.options.display.max_rows=500
pd.options.display.max_columns=50
```

```python
train = pd.read_feather('../data/input/train_data.feather')
oof = np.load('../logs/lgbm_reg02_20191219153718_23.501/oof.npy')
train['oof'] = oof
train['ae'] = np.abs(train['salary'] - train['oof'])
```

```python
df = train[train['area'] == '沖縄県']
```

```python
plt.figure(figsize=(16, 5))
plt.plot(df.sort_values(by='salary', ascending=False)['ae'], 'o')
```
