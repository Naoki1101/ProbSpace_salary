```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.options.display.max_rows=500
pd.options.display.max_columns=50
```

```python
train = pd.read_feather('../data/input/train_data.feather')
train['salary_level'] = (train['salary'] // 100).astype(int)
```

```python
all_area = [
    '北海道', 
    '青森県', '岩手県', '宮城県', '秋田県', '山形県', '福島県', 
    '茨城県', '栃木県', '群馬県', '埼玉県', '千葉県', '東京都', '神奈川県', 
    '新潟県', '富山県', '石川県', '福井県', '山梨県', '長野県', '岐阜県', '静岡県', '愛知県', '三重県', '滋賀県', 
    '京都府', '大阪府', '兵庫県', '奈良県', '和歌山県', 
    '鳥取県', '島根県', '岡山県', '広島県', '山口県', 
    '徳島県', '香川県', '愛媛県', '高知県', 
    '福岡県', '佐賀県', '長崎県', '熊本県', '大分県', '宮崎県', '鹿児島県', '沖縄県'
]
```

# salary

```python
plt.figure(figsize=(16, 30))
for i, area in enumerate(all_area):
    plt.subplot(10, 5, i + 1)
    df = train[train['area'] == area]
    sns.countplot(df['salary_level'])
    plt.title(area)
    plt.xlim([-0.5, 10.5])
plt.tight_layout()
```

# position

```python
plt.figure(figsize=(16, 30))
for i, area in enumerate(all_area):
    plt.subplot(10, 5, i + 1)
    df = train[train['area'] == area]
    sns.countplot(df['position'])
    plt.title(area)
    plt.xlim([-0.5, 4.5])
plt.tight_layout()
```
